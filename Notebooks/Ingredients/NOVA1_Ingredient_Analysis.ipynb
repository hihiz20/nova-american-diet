# NOVA 1: Unprocessed or Minimally Processed Foods

**NOVA 1 foods** are whole foods that can be eaten without the use of synthetic or lab-made chemicals. Provided by nature, they supply humans with essential nutrients like vitamins, protein, fiber, amino acids, folic acid, and fat (Monteiro et al., 2018; Monteiro et al., 2019).

### Essential Nutrients & Examples
* **Examples:** Fruits, plants, vegetables, seeds, fresh salmon, fresh eggs, seaweed, and water.
* **Nutrients:** Vitamins, protein, fiber, amino acids, folic acid, and fat.

### What is Minimal Processing?
Minimal processing is a method designed to **physically modify** natural food (Monteiro et al., 2018; Monteiro et al., 2019). 
* Done by hand, tools, or machines without adding any synthetic or lab-made chemicals.
* **No** added sugar, salt, syrup, fats, or oil.
* **Examples:** Cleaning, removing unwanted parts, crushed chili, boiled eggs, filtering, dried fruits, pasteurization, and freezing.

---

| Category | Details / Examples |
| :--- | :--- |

## Reviewing NOVA 1 Categories

There are 20 unique food categories currently classified as NOVA 1, but not 
all truly belong here. After reviewing each category against the NOVA 1 
definition (Monteiro et al., 2019) — which states that NOVA 1 foods must 
have **no added salt, sugar, oils, or fats** — the following categories 
need to be reclassified:

### Reclassify to NOVA 3 (Processed)

- **Pickles, Olives, Peppers & Relishes** (44,213 products)  
  Pickling is a preservation method that involves submerging food in vinegar 
  or salt brine (Britannica, 2026). The standard ingredients used in 
  commercial pickling include sodium chloride, sugar, citric acid, and 
  vinegar, with additional preservatives such as calcium chloride often 
  added (Fleming et al., 1993). This was confirmed by the USDA ingredient 
  data, where 73.84% of products in this category contained added salt 
  and 26.28% contained vinegar.

- **Tomatoes** (14,510 products)  
  The majority are canned tomatoes containing added salt and citric acid. 
  Fresh tomatoes would be NOVA 1, but the ingredient lists confirm these 
  are processed.

### Reclassify to NOVA 2 (Culinary Ingredient)

- **Herbs/Spices/Extracts** (1,173 products)  
  "Extracts" indicates processing beyond minimal (e.g., vanilla extract 
  requires alcohol and industrial extraction).

### Categories that remain NOVA 1

- **Herbs & Spices** - plain dried/ground herbs = minimal processing
- All **"Unprepared/Unprocessed"** categories - by definition NOVA 1
- **Peppers**, **Chickpeas**, **Berries/Small Fruit** are whole foods
  
### Impact

This reclassification moves approximately 59,896 products (44,213 + 14,510 + 1,173) 
out of NOVA 1, reducing it from 77,038 to approximately 17,142 products — a 
significant correction that demonstrates why category-level classification alone 
is insufficient without ingredient-level validation.

```python
# Although tomatoes looks like NOVA 1 classification, I assume most are 
# NOT considered NOVA 1 due to added ingredients. I can check my assumption

# Reclassify tomatoes with added ingredients to NOVA 3
tomato_mask = (df["branded_food_category"] == "Tomatoes") & \
              (df["ingredients"].str.lower().str.contains(
                  "salt|citric acid|sugar|oil", na=False
              ))

df.loc[tomato_mask, "nova_category"] = "NOVA 3 - Processed"
print(f"Tomatoes moved to NOVA 3: {tomato_mask.sum():,}")
print(f"Tomatoes remaining in NOVA 1: {(df['branded_food_category'] == 'Tomatoes').sum() - tomato_mask.sum():,}")

# Check NOVA 1 count now
nova1_check = df[df["nova_category"] == "NOVA 1 - Unprocessed/Minimally Processed"]
print(f"\nNOVA 1 count: {len(nova1_check):,}")
```

```python
# Since there are 831 tomatoes without added ingredients. Will keep it.
# And the rest will be saved in a file and insert it to NOVA 3 in the future.
# Saving reclassified products before moving them to maintaining data integrity

# Pickles moves to NOVA 3
pickles_to_nova3 = df[df["branded_food_category"] == "Pickles, Olives, Peppers & Relishes"].copy()
pickles_to_nova3.to_csv("reclassified_nova1_to_nova3_pickles.csv", index=False)
print(f"Saved pickles: {len(pickles_to_nova3):,} products")

# Tomatoes with added ingredients moves to NOVA 3
tomatoes_to_nova3 = df[(df["branded_food_category"] == "Tomatoes") & 
                       (df["ingredients"].str.lower().str.contains(
                           "salt|citric acid|sugar|oil", na=False
                       ))].copy()
tomatoes_to_nova3.to_csv("reclassified_nova1_to_nova3_tomatoes.csv", index=False)
print(f"Saved tomatoes: {len(tomatoes_to_nova3):,} products")

# Herbs/Spices/Extracts to NOVA 2
extracts_to_nova2 = df[df["branded_food_category"] == "Herbs/Spices/Extracts"].copy()
extracts_to_nova2.to_csv("reclassified_nova1_to_nova2_extracts.csv", index=False)
print(f"Saved extracts: {len(extracts_to_nova2):,} products")

# Herbs & Spices with added ingredients to NOVA 3
herbs_to_nova3 = df[(df["branded_food_category"].isin(["Herbs & Spices", "Herbs/Spices (Shelf Stable)"])) & 
                    (df["ingredients"].str.lower().str.contains(
                        "salt|citric acid|sugar|oil|vinegar", na=False
                    ))].copy()
herbs_to_nova3.to_csv("reclassified_nova1_to_nova3_herbs.csv", index=False)
print(f"Saved herbs with additives: {len(herbs_to_nova3):,} products")
```

```python
# Reclassify products with added ingredients across remaining categories
check_categories = [
    "Herbs & Spices",
    "Herbs/Spices (Shelf Stable)"
]

for cat in check_categories:
    mask = (df["branded_food_category"] == cat) & \
           (df["ingredients"].str.lower().str.contains(
               "salt|citric acid|sugar|oil|vinegar", na=False
           ))
    df.loc[mask, "nova_category"] = "NOVA 3 - Processed"
    print(f"{cat}: moved {mask.sum():,} products to NOVA 3")
```

```python
# Pickles — entire category to NOVA 3 (already confirmed by research)
df.loc[df["branded_food_category"] == "Pickles, Olives, Peppers & Relishes", 
       "nova_category"] = "NOVA 3 - Processed"

# Herbs/Spices/Extracts — entire category to NOVA 2 (extracts = industrial processing)
df.loc[df["branded_food_category"] == "Herbs/Spices/Extracts", 
       "nova_category"] = "NOVA 2 - Culinary Ingredient"
```

```python
# Final NOVA 1 check
nova1_final = df[df["nova_category"] == "NOVA 1 - Unprocessed/Minimally Processed"]
print(f"\nFinal NOVA 1 count: {len(nova1_final):,}")
print(f"\nRemaining NOVA 1 categories:")
print(nova1_final["branded_food_category"].value_counts())
```

## Ingredient-Level Validation of NOVA 1

The category-level classification initially placed 77,038 products in NOVA 1. 
After reclassifying obvious category errors (pickles, canned tomatoes), the 
count dropped to 13,754. However, analyzing the actual ingredients of these 
remaining products revealed that many still contained additives, preservatives, 
and chemically modified ingredients that disqualify them from NOVA 1.

**Key finding:** The category label is not enough — a product labeled 
"Unprepared/Unprocessed" may still contain added salt, sodium tripolyphosphate, 
high fructose corn syrup, or artificial preservatives. The ingredients tell 
the truth, not the category name.

### Disqualifying Ingredients Found in "NOVA 1" Products

| Reclassify to | Ingredients Found | Examples |
|---------------|-------------------|----------|
| **NOVA 3** (Processed) | Added salt, sugar, vinegar, oils | Salt (20.91%), sugar (8.13%), vinegar (7.39%) |
| **NOVA 4** (Ultra-Processed) | Chemical additives, preservatives, modified ingredients | Sodium tripolyphosphate (4.95%), maltodextrin (2.58%), HFCS (1.55%), sodium nitrite (1.17%) |

### Why This Matters

This validates that **ingredient-level analysis is essential** for accurate NOVA 
classification. Category names assigned by the USDA describe what a product 
*is*, not how it was *made*. A product in the "Meat Unprepared/Unprocessed" 
category may still be injected with sodium phosphate solutions and preserved 
with sodium nitrite — making it NOVA 3 or NOVA 4, not NOVA 1.

> "The category is innocent, but the ingredients tell the truth." - Hizkia Permata

```python
# Re-filter NOVA 1 with the corrected classification
df_nova1 = df[df["nova_category"] == "NOVA 1 - Unprocessed/Minimally Processed"].copy()

print(f"NOVA 1 products: {len(df_nova1):,}")

# Split all ingredients and count occurrences
all_ingredients = []

for item in df_nova1["ingredients"].dropna():
    parts = [ing.strip().upper() for ing in item.split(",")]
    parts = [p for p in parts if p]
    all_ingredients.extend(parts)

# Count and rank
from collections import Counter
ingredient_freq = Counter(all_ingredients)
top_50 = pd.DataFrame(ingredient_freq.most_common(50), 
                       columns=["ingredient", "count"])

# Add percentage
top_50["% of NOVA 1"] = (top_50["count"] / len(df_nova1) * 100).round(2)

print(f"\nTotal unique ingredients found: {len(ingredient_freq):,}")
top_50
```

```python
# Save audit trail FIRST — products about to be reclassified from NOVA 1
# based on ingredient-level analysis

nova1_current = df[df["nova_category"] == "NOVA 1 - Unprocessed/Minimally Processed"].copy()
nova1_current.to_csv("nova1_before_ingredient_reclassification.csv", index=False)
print(f"Saved pre-reclassification NOVA 1: {len(nova1_current):,} products")
```

```python
nova4_markers = [
    "potassium chloride", "sodium tripolyphosphate", "natural flavor",
    "natural flavors", "natural flavoring", "sodium phosphate",
    "modified food starch", "modified corn starch", "sodium bisulfite",
    "maltodextrin", "sodium erythorbate", "high fructose corn syrup",
    "caramel color", "sodium nitrite", "hydrolyzed corn protein",
    "citric acid", "basted"
]

nova3_markers = [
    "salt", "sugar", "vinegar", "dextrose", "corn syrup",
    "soybean oil", "canola oil", "sunflower oil"
]

# Save products about to move
nova4_to_move = df[(df["nova_category"] == "NOVA 1 - Unprocessed/Minimally Processed") & 
                   (df["ingredients"].str.lower().str.contains(
                       "|".join(nova4_markers), na=False
                   ))].copy()

nova3_to_move = df[(df["nova_category"] == "NOVA 1 - Unprocessed/Minimally Processed") & 
                   (~df.index.isin(nova4_to_move.index)) &
                   (df["ingredients"].str.lower().str.contains(
                       "|".join(nova3_markers), na=False
                   ))].copy()

# Save audit files
nova4_to_move.to_csv("reclassified_nova1_to_nova4_ingredients.csv", index=False)
nova3_to_move.to_csv("reclassified_nova1_to_nova3_ingredients.csv", index=False)
print(f"Saved NOVA 1 → NOVA 4: {len(nova4_to_move):,} products")
print(f"Saved NOVA 1 → NOVA 3: {len(nova3_to_move):,} products")

# Apply reclassification
df.loc[nova4_to_move.index, "nova_category"] = "NOVA 4 - Ultra-Processed"
df.loc[nova3_to_move.index, "nova_category"] = "NOVA 3 - Processed"

# Save immediately
df.to_csv("NOVA1_ingredients_result.csv", index=False)

# Verify
nova1_after = df[df["nova_category"] == "NOVA 1 - Unprocessed/Minimally Processed"]
print(f"\nNOVA 1 before: 13,754")
print(f"NOVA 1 after: {len(nova1_after):,}")
print(f"Moved to NOVA 4: {len(nova4_to_move):,}")
print(f"Moved to NOVA 3: {len(nova3_to_move):,}")
```

```python
# What remains in NOVA 1?
nova1_final = df[df["nova_category"] == "NOVA 1 - Unprocessed/Minimally Processed"]

print(f"Final NOVA 1 count: {len(nova1_final):,}")
print(f"\nRemaining categories:")
print(nova1_final["branded_food_category"].value_counts())

# Re-run ingredient analysis on cleaned NOVA 1
all_ingredients = []
for item in nova1_final["ingredients"].dropna():
    parts = [ing.strip().upper() for ing in item.split(",")]
    parts = [p for p in parts if p]
    all_ingredients.extend(parts)

from collections import Counter
top_20 = pd.DataFrame(Counter(all_ingredients).most_common(20),
                       columns=["ingredient", "count"])
top_20["% of NOVA 1"] = (top_20["count"] / len(nova1_final) * 100).round(2)

print(f"\nTop 20 ingredients in cleaned NOVA 1:")
top_20
```

## NOVA 1 — Final Results After Ingredient-Level Validation

After systematic ingredient analysis, NOVA 1 was reduced from 77,038 to 
**9,641 products** - an 87.5% reduction. Only **0.5%** of all US branded 
food products are truly unprocessed or minimally processed.

The top 20 ingredients in the validated NOVA 1 category are exclusively 
whole foods: pork, garlic, black pepper, cumin, cinnamon, basil, snapper, 
tomatoes, and other natural herbs and spices. No added salt, sugar, 
preservatives, or chemical additives remain.

This confirms that category-level classification alone is insufficient - 
ingredient-level validation is essential for accurate NOVA classification.

## Important Context for the 0.5% Finding

The USDA Branded Foods database tracks packaged, labeled products - not 
all food available in the US. Fresh produce, bulk grains, and unpackaged 
items sold without ingredient labels are not represented in this dataset.

Therefore, the 0.5% figure represents the proportion of **branded, 
packaged food products** that are truly minimally processed - not the 
proportion of all food Americans consume. The low percentage reflects 
the reality that most packaged food requires some form of added 
ingredient for preservation, shelf stability, or regulatory compliance.

Additionally, some products may contain ingredients added for food 
safety purposes (e.g., sodium phosphate in poultry for moisture 
retention) rather than for taste or processing - a nuance that 
keyword-based classification cannot distinguish.
